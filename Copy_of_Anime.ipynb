```python
def display_missing_values(df, name):
    """Displays missing values in a DataFrame in a clean, readable format."""
    missing = df.isnull().sum().to_frame(name="Missing Values")

    print(f"\n{name} Dataset:")
    display(missing.style.set_properties(**{
        "border": "1px solid black",
        "text-align": "center"
    }).set_table_styles([{"selector": "th",
"props": [("background-color", "#f4f4f4"), ("font-weight", "bold")]
    }]))
print("üîç Missing Values in Each Dataset:")
datasets = {
    "Anime": anime_df,
    "Ratings": ratings_df,
    "Test": test_df
}

for name, df in datasets.items():
    display_missing_values(df, name)

```

<h4 style="color: #00eaff; font-family: 'Rajdhani', sans-serif; font-style: italic;">

</h4>


```python
anime_df['rating'] = anime_df['rating'].fillna(anime_df.groupby('type')['rating'].transform('mean'))
anime_df['genre'] = anime_df['genre'].fillna('Unknown')
anime_df['name'] = anime_df['name'].fillna('Unknown')
```

<h4 style="color: #00eaff; font-family: 'Rajdhani', sans-serif; font-style: italic;">
    Replacing Nulls  
</h4>


```python
#Replace Null-genre entries
anime_df['genre'] = anime_df.apply(lambda row: 'Comedy' if row['type'] == 'Movie' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
anime_df['genre'] = anime_df.apply(lambda row: 'Comedy' if row['type'] == 'TV' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
anime_df['genre'] = anime_df.apply(lambda row: 'Hentai' if row['type'] == 'OVA' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
anime_df['genre'] = anime_df.apply(lambda row: 'Comedy' if row['type'] == 'Special' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
anime_df['genre'] = anime_df.apply(lambda row: 'Music' if row['type'] == 'Music' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
anime_df['genre'] = anime_df.apply(lambda row: 'Comedy' if row['type'] == 'ONA' and (not row['genre'] or pd.isna(row['genre'])) else row['genre'], axis=1)
```

```python
#Replace Null-type entries
anime_df['type'] = anime_df['type'].replace('', 'TV').fillna('TV')
```

```python
#Replace Null-ratings entries
average_rating_per_type = anime_df.groupby('type')['rating'].transform('mean')
anime_df['rating'] = anime_df['rating'].fillna(average_rating_per_type)

```

<h4 style="color: #00eaff; font-family: 'Rajdhani', sans-serif; font-style: italic;">
    üïµÔ∏è Identify Active and Inactive Users  
</h4>


```python
# Calculate the number of ratings per user
user_ratings_count = ratings_df.groupby('user_id')['rating'].count()
active_threshold = user_ratings_count.quantile(0.75)  # 75th percentile
inactive_threshold = user_ratings_count.quantile(0.25)  # 25th percentile
active_users = user_ratings_count[user_ratings_count > active_threshold].index
inactive_users = user_ratings_count[user_ratings_count < inactive_threshold].index
print(f"Number of active users: {len(active_users)}")
print(f"Number of inactive users: {len(inactive_users)}")
```

We have 69,481 total users, but only 34,614 (17,326 active + 17,288 inactive) are accounted for
in our user activity analysis.

<h4 style="color: #00eaff; font-family: 'Rajdhani', sans-serif; font-style: italic;">
    üìä Categorizing Users Based on Activity Levels and Visualizing the Distribution  
</h4>


```python

user_ratings_count = ratings_df.groupby('user_id')['rating'].count()
very_active_threshold = user_ratings_count.quantile(0.9)  # Top 10%
active_threshold = user_ratings_count.quantile(0.75)      # Top 25%
inactive_threshold = user_ratings_count.quantile(0.25)    # Bottom 25%
very_inactive_threshold = user_ratings_count.quantile(0.1)  # Bottom 10%
user_activity = pd.DataFrame({'user_id': user_ratings_count.index})
user_activity['user_type'] = np.where(
    user_ratings_count > very_active_threshold, 'very_active',
    np.where(user_ratings_count > active_threshold, 'active',
             np.where(user_ratings_count < inactive_threshold, 'inactive', 'low_activity'))
)

print("User Type Distribution:")
print(user_activity['user_type'].value_counts())
user_type_order = ['low_activity', 'inactive', 'active', 'very_active']
plt.figure(figsize=(10, 6))
sns.countplot(x='user_type', data=user_activity, order=user_type_order, palette='pastel')
plt.title('Distribution of User Activity Levels')
plt.xlabel('User Type')
plt.ylabel('Count of Users')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# !pip install mlflow
# !pip install scikit-surprise
import pandas as pd
import numpy as np
import mlflow
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split as surprise_train_test_split

# --- Preprocessing (Updated to Fix Failures) ---
df = merged_df.copy()

# Scale numeric columns
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
df['episodes'] = np.log1p(df['episodes'])
df['members'] = np.log1p(df['members'])
numeric_cols = ['episodes', 'members']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Scale the target variable (user_rating)
scaler = MinMaxScaler(feature_range=(0, 1))
df['scaled_score'] = scaler.fit_transform(df[['user_rating']])

# Create feature matrix X and target variable y
X = df[['user_id', 'anime_id']].values
y = df["scaled_score"].values
test_set_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_set_size, random_state=73)
X_train_array = [X_train[:, 0], X_train[:, 1]]
X_test_array = [X_test[:, 0], X_test[:, 1]]

# Function to preprocess features with weighting
def preprocess_features(df):
    df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
    bins = [0, 1, 12, 50, float('inf')]
    labels = ['single', 'short', 'medium', 'long']
    df['episode_bin'] = pd.cut(df['episodes'], bins=bins, labels=labels, include_lowest=True)

    # Weight genres higher by repeating them
    df['weighted_genre'] = df['genre'].fillna("").astype(str).apply(lambda x: ' '.join([x] * 3))  # 3x weight
    df['weighted_type'] = df['type'].fillna("").astype(str).apply(lambda x: ' '.join([x] * 2))    # 2x weight
    df['weighted_episode_bin'] = df['episode_bin'].astype(str).replace('nan', '')

    combined_features = df.apply(
        lambda x: f"{x['weighted_genre']} {x['weighted_type']} {x['weighted_episode_bin']}", axis=1
    )
    return combined_features

rec_data = anime_df.drop_duplicates(subset="name", keep="first").reset_index(drop=True)

# Scale members for similarity weighting
scaler = MinMaxScaler()
rec_data['members_scaled'] = scaler.fit_transform(rec_data[['members']])

# Compute TF-IDF matrix with weighted features
combined_features = preprocess_features(rec_data)
tfv = TfidfVectorizer(
    min_df=3,
    max_features=None,
    strip_accents="unicode",
    analyzer="word",
    token_pattern=r"\S+",
    ngram_range=(1, 3),
    stop_words="english",
)
tfidf_matrix = tfv.fit_transform(combined_features)
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Combine cosine similarity with members similarity
def weighted_similarity(cosine_sim, members_scaled, alpha=0.8):
    members_sim = 1 / (1 + np.abs(members_scaled.reshape(-1, 1) - members_scaled.reshape(1, -1)))
    return alpha * cosine_sim + (1 - alpha) * members_sim

cosine_sim_combined = weighted_similarity(cosine_sim, rec_data['members_scaled'].values, alpha=0.8)

# Encode user and anime IDs
user_encoder = LabelEncoder()
df["user_encoded"] = user_encoder.fit_transform(df["user_id"])
num_users = len(user_encoder.classes_)
anime_encoder = LabelEncoder()
df["anime_encoded"] = anime_encoder.fit_transform(df["anime_id"])
num_animes = len(anime_encoder.classes_)

# --- Collaborative Filtering Setup ---
# Prepare Surprise dataset
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(ratings_df[['user_id', 'anime_id', 'rating']], reader)
trainset, testset = surprise_train_test_split(data, test_size=0.2, random_state=42)

# Initialize MLflow
mlflow.set_experiment("Anime_Recommendation_Collaborative_2025")
```

```python
indices = pd.Series(rec_data.index, index=rec_data['name'])

def give_rec(title, cosine_sim=cosine_sim_combined, top_n=10):
    if title not in indices:
        return f"‚ùå '{title}' not found in the dataset!"

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sorted_scores = sorted(sim_scores, key=lambda x: (x[1], rec_data.iloc[x[0]]['rating']), reverse=True)
    top_indices = [i[0] for i in sorted_scores if i[0] != idx][:top_n]

    num_recs = len(top_indices)
    if num_recs == 0:
        return "No recommendations available."

    return pd.DataFrame({
        "No": range(1, num_recs + 1),
        "Anime Name": rec_data["name"].iloc[top_indices].values,
        "Genre": rec_data["genre"].iloc[top_indices].values,
        "Rating": rec_data["rating"].iloc[top_indices].values
    }).set_index("No")

# Example usage
recommendations = give_rec("Naruto")
print(recommendations)


```

```python
def get_user_content_recommendations(user_id, ratings_df, cosine_sim, top_n=10, min_rating=8):
    """
    Get content-based recommendations for a user based on their highly rated anime.
    """
    user_ratings = ratings_df[ratings_df['user_id'] == user_id]
    liked_anime = user_ratings[user_ratings['rating'] >= min_rating]

    if liked_anime.empty:
        return f"No highly rated anime (rating >= {min_rating}) found for user {user_id}."

    liked_anime_ids = liked_anime['anime_id'].values
    liked_titles = rec_data[rec_data['anime_id'].isin(liked_anime_ids)]['name'].values

    all_recs = []
    for title in liked_titles:
        recs = give_rec(title, cosine_sim, top_n=top_n)
        if isinstance(recs, str):
            continue
        # Convert DataFrame to list of (title, score) tuples
        for _, row in recs.iterrows():
            # Estimate score as similarity (not directly available, use rating as proxy)
            all_recs.append((row['Anime Name'], row['Rating'] / 10))  # Normalize rating to [0, 1]

    if not all_recs:
        return f"No recommendations could be generated for user {user_id}."

    rec_df = pd.DataFrame(all_recs, columns=['title', 'score'])
    rec_df = rec_df.groupby('title').agg({'score': 'mean'}).reset_index()
    rec_df = rec_df.sort_values('score', ascending=False).head(top_n)
    return list(zip(rec_df['title'], rec_df['score']))
```

```python
# Function to predict rating for RMSE evaluation
def predict_rating(user_id, anime_id, ratings_df, cosine_sim, rec_data, min_rating=8):
    target_title = rec_data[rec_data['anime_id'] == anime_id]['name'].values
    if len(target_title) == 0:
        return ratings_df['rating'].mean()

    target_title = target_title[0]
    user_ratings = ratings_df[ratings_df['user_id'] == user_id]
    liked_anime = user_ratings[user_ratings['rating'] >= min_rating]
    if liked_anime.empty:
        return ratings_df['rating'].mean()

    target_idx = indices[target_title]
    liked_anime_ids = liked_anime['anime_id'].values
    liked_indices = rec_data[rec_data['anime_id'].isin(liked_anime_ids)].index
    if len(liked_indices) == 0:
        return ratings_df['rating'].mean()

    sim_scores = cosine_sim[target_idx, liked_indices]
    user_ratings = liked_anime[liked_anime['anime_id'].isin(rec_data.loc[liked_indices, 'anime_id'])]['rating'].values
    if len(sim_scores) == 0 or len(user_ratings) == 0:
        return ratings_df['rating'].mean()

    weighted_rating = np.sum(sim_scores * user_ratings) / (np.sum(sim_scores) + 1e-8)
    return weighted_rating
```

```python
# --- Utility Matrix for Collaborative Filtering ---
print("Creating utility matrix...")
# Further reduce sample size for utility matrix
sample_size_utility = 100000  # Reduced to 100K ratings
ratings_utility = ratings_subset.sample(n=sample_size_utility, random_state=42)

utility_matrix = ratings_utility.pivot_table(
    index='user_id',
    columns='anime_id',
    values='rating',
    fill_value=0
)

# Normalize ratings
user_means = utility_matrix.replace(0, np.nan).mean(axis=1)
utility_matrix_normalized = utility_matrix.sub(user_means, axis=0).fillna(0)

# Drop users with no ratings
utility_matrix = utility_matrix.loc[utility_matrix.sum(axis=1) != 0]

# Convert to sparse matrix
utility_matrix_sparse = csr_matrix(utility_matrix)

# Compute user-user similarity
print("Computing user-user similarity...")
user_similarity = cosine_similarity(utility_matrix_sparse)
user_similarity_df = pd.DataFrame(user_similarity, index=utility_matrix.index, columns=utility_matrix.index)

# Compute item-item similarity
print("Computing item-item similarity...")
utility_matrix_t = utility_matrix_normalized.T
utility_matrix_t = utility_matrix_t.loc[utility_matrix_t.sum(axis=1) != 0]
utility_matrix_t_sparse = csr_matrix(utility_matrix_t)
item_similarity = cosine_similarity(utility_matrix_t_sparse)
item_similarity_df = pd.DataFrame(item_similarity, index=utility_matrix_t.index, columns=utility_matrix_t.index)

# Clear memory
del utility_matrix_normalized, utility_matrix_t, utility_matrix_sparse, utility_matrix_t_sparse
gc.collect()
```

```python
# --- Collaborative Filtering Recommender Functions ---
def get_top_n_recommendations_user_based(user_id, user_similarity_df, utility_matrix, k=5, n=10):
    if user_id not in user_similarity_df.index:
        # Cold-start: Recommend popular items
        popular_items = utility_matrix.mean().sort_values(ascending=False).head(n).index.tolist()
        return popular_items

    sim_scores = user_similarity_df.loc[user_id]
    similar_users = sim_scores.sort_values(ascending=False).index[1:k+1]

    top_rated_items = {}
    for similar_user in similar_users:
        user_ratings = utility_matrix.loc[similar_user]
        user_rated_items = user_ratings[user_ratings > 0].index
        reference_rated_items = utility_matrix.loc[user_id][utility_matrix.loc[user_id] > 0].index
        unrated_items = [item for item in user_rated_items if item not in reference_rated_items]

        for item in unrated_items:
            rating = user_ratings[item]
            if item not in top_rated_items:
                top_rated_items[item] = []
            top_rated_items[item].append(rating)

    item_scores = [(item, np.mean(ratings)) for item, ratings in top_rated_items.items()]
    item_scores = sorted(item_scores, key=lambda x: x[1], reverse=True)
    return [item for item, score in item_scores[:n]]
```

```python
def get_top_n_recommendations_item_based(anime_id, item_similarity_df, utility_matrix, k=5, n=10):
    if anime_id not in item_similarity_df.index:
        # Cold-start: Recommend popular items
        popular_items = utility_matrix.mean().sort_values(ascending=False).head(n).index.tolist()
        return popular_items

    sim_scores = item_similarity_df.loc[anime_id]
    similar_items = sim_scores.sort_values(ascending=False).index[1:k+1]
    return similar_items.tolist()[:n]
```

```python
def predict_rating_user_based(user_id, anime_id, user_similarity_df, utility_matrix, k=5, threshold=0.1):
    if user_id not in user_similarity_df.index or anime_id not in utility_matrix.columns:
        # Cold-start: Return global mean rating
        return ratings_utility['rating'].mean()

    sim_scores = user_similarity_df.loc[user_id]
    similar_users = sim_scores.sort_values(ascending=False).index[1:k+1]

    sim_users_ratings = []
    sim_values = []
    for similar_user in similar_users:
        sim = sim_scores[similar_user]
        if sim < threshold:
            continue
        rating = utility_matrix.loc[similar_user, anime_id]
        if rating > 0:
            sim_users_ratings.append(rating)
            sim_values.append(sim)

    if not sim_users_ratings:
        return ratings_utility['rating'].mean()

    weighted_sum = sum(r * s for r, s in zip(sim_users_ratings, sim_values))
    sim_sum = sum(sim_values)
    return weighted_sum / sim_sum
```

```python
def predict_rating_item_based(user_id, anime_id, item_similarity_df, utility_matrix, k=5, threshold=0.1):
    if anime_id not in item_similarity_df.index or user_id not in utility_matrix.index:
        # Cold-start: Return global mean rating
        return ratings_utility['rating'].mean()

    user_ratings = utility_matrix.loc[user_id]
    rated_items = user_ratings[user_ratings > 0].index

    sim_items_ratings = []
    sim_values = []
    for rated_item in rated_items:
        if rated_item not in item_similarity_df.index:
            continue
        sim = item_similarity_df.loc[anime_id, rated_item]
        if sim < threshold:
            continue
        rating = user_ratings[rated_item]
        sim_items_ratings.append(rating)
        sim_values.append(sim)

    if not sim_items_ratings:
        return ratings_utility['rating'].mean()

    weighted_sum = sum(r * s for r, s in zip(sim_items_ratings, sim_values))
    sim_sum = sum(sim_values)
    return weighted_sum / sim_sum
```

```python
# --- Evaluate Diversity and Hit Rate ---
def compute_genre_diversity(recommendations, rec_data):
    if isinstance(recommendations, str):
        return 0
    genres = []
    for anime_id in recommendations:
        if anime_id in rec_data['anime_id'].values:
            genre = rec_data[rec_data['anime_id'] == anime_id]['genre'].iloc[0]
            genres.extend(genre.split(', ') if isinstance(genre, str) else [])
    unique_genres = len(set(genres))
    return unique_genres

def compute_hit_rate(user_id, recommendations, utility_matrix, threshold=8):
    if isinstance(recommendations, str):
        return 0
    user_high_ratings = utility_matrix.loc[user_id]
    user_high_rated_items = user_high_ratings[user_high_ratings >= threshold].index.tolist()
    hits = sum(1 for item in recommendations if item in user_high_rated_items)
    return hits / len(recommendations) if recommendations else 0

# User-Based
diversity_score_user = compute_genre_diversity(recs_user, rec_data)
print(f"\nGenre Diversity of User-Based Recommendations: {diversity_score_user} unique genres")

hit_rate_user = compute_hit_rate(sample_user, recs_user, utility_matrix)
print(f"Hit Rate for User-Based Recommendations for User {sample_user} (threshold=8): {hit_rate_user:.2f}")

# Item-Based
user_rated_items = utility_matrix.loc[sample_user][utility_matrix.loc[sample_user] > 0].index
recs_item_based_user = []
for anime_id in user_rated_items[:5]:
    similar_items = get_top_n_recommendations_item_based(anime_id, item_similarity_df, utility_matrix, k=5, n=5)
    recs_item_based_user.extend(similar_items)
recs_item_based_user = list(set(recs_item_based_user))[:5]

diversity_score_item = compute_genre_diversity(recs_item_based_user, rec_data)
print(f"Genre Diversity of Item-Based Recommendations: {diversity_score_item} unique genres")

hit_rate_item = compute_hit_rate(sample_user, recs_item_based_user, utility_matrix)
print(f"Hit Rate for Item-Based Recommendations for User {sample_user} (threshold=8): {hit_rate_item:.2f}")
```

```python
# --- Evaluate Utility Matrix Collaborative Filtering ---
print("\nEvaluating Utility Matrix Collaborative Filtering (User-Based)...")
testset_list = list(testset)
np.random.seed(42)
sampled_testset = np.random.choice(testset_list, size=2000, replace=False)  # Further reduced for speed

collab_user_predictions = []
actuals = []
for uid, iid, true_r in sampled_testset:
    pred = predict_rating_user_based(uid, iid, user_similarity_df, utility_matrix)
    collab_user_predictions.append(pred)
    actuals.append(true_r)

collab_user_rmse = np.sqrt(mean_squared_error(actuals, collab_user_predictions))
print(f"Utility Matrix User-Based Collaborative Filtering RMSE (Sampled): {collab_user_rmse:.4f}")

print("\nEvaluating Utility Matrix Collaborative Filtering (Item-Based)...")
collab_item_predictions = []
for uid, iid, true_r in sampled_testset:
    pred = predict_rating_item_based(uid, iid, item_similarity_df, utility_matrix)
    collab_item_predictions.append(pred)

collab_item_rmse = np.sqrt(mean_squared_error(actuals, collab_item_predictions))
print(f"Utility Matrix Item-Based Collaborative Filtering RMSE (Sampled): {collab_item_rmse:.4f}")

with mlflow.start_run(run_name="UtilityMatrix_Collab_Model"):
    mlflow.log_metric("user_based_rmse_sampled", collab_user_rmse)
    mlflow.log_metric("item_based_rmse_sampled", collab_item_rmse)

# --- Hybrid Model (Surprise + Utility Matrix) ---
print("\nEvaluating Hybrid Model...")
hybrid_predictions = []
actuals = []
for uid, iid, true_r in sampled_testset:
    collab_surprise_pred = best_collab_model.predict(uid, iid).est
    collab_user_pred = predict_rating_user_based(uid, iid, user_similarity_df, utility_matrix)
    collab_item_pred = predict_rating_item_based(uid, iid, item_similarity_df, utility_matrix)
    weight_surprise = 1 / best_rmse
    weight_user = 1 / collab_user_rmse
    weight_item = 1 / collab_item_rmse
    hybrid_pred = (weight_surprise * collab_surprise_pred +
                   weight_user * collab_user_pred +
                   weight_item * collab_item_pred) / (weight_surprise + weight_user + weight_item)
    hybrid_predictions.append(hybrid_pred)
    actuals.append(true_r)

hybrid_rmse = np.sqrt(mean_squared_error(actuals, hybrid_predictions))
print(f"Hybrid Model RMSE (Sampled): {hybrid_rmse:.4f}")

# Log to MLflow
with mlflow.start_run(run_name="Hybrid_Model"):
    mlflow.log_param("collaborative_model_surprise", best_model_name)
    mlflow.log_param("collab_surprise_rmse", best_rmse)
    mlflow.log_param("collab_user_rmse", collab_user_rmse)
    mlflow.log_param("collab_item_rmse", collab_item_rmse)
    mlflow.log_metric("hybrid_rmse_sampled", hybrid_rmse)

# Clear memory
del user_similarity_df, item_similarity_df, utility_matrix, hybrid_predictions, actuals
gc.collect()
```
